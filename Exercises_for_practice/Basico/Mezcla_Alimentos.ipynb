# Mezcla de Alimentos
Una granja desea crear una mezcla de alimento para animales a partir de 2 fuentes, F1 y F2.
## Fuentes y Costos

* F1: Cuesta 2\$ euros por unidad. Contiene 3g de Nutriente A y 2g de Nutriente B.
* F2: Cuesta 3\$ euros por unidad. Contiene 1g de Nutriente A y 4g de Nutriente B.

## Requisitos Mínimos

* Se necesitan al menos 10g de Nutriente A.
* Se necesitan al menos 12g de Nutriente B.

```python
import gurobipy as gp

m=gp.Model("Mezcla de Alimentos")

# Variables
# x1 = unidades de F1
# x2 = unidades de F2
x1 = m.addVar(name="x1_Fuente1", lb=0)
x2 = m.addVar(name="x2_Fuente2", lb=0)

# Objetivo: Minimizar Costo
m.setObjective(2*x1 + 3*x2, gp.GRB.MINIMIZE)

# Restricciones
m.addConstr(3*x1+1*x2 >=10, "Req_NutrienteA")
m.addConstr(2*x1+4*x2>=12, "Req_NutrienteB")

m.optimize()

if m.status==gp.GRB.OPTIMAL:
    print(f"Costo Mínimo: {m.objVal:.2f} €")
    print(f"Usar F1: {x1.X:.2f} unidades")
    print(f"Usar F2: {x2.X:.2f} unidades")
```
