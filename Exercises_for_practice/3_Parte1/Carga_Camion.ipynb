# Carga de un camión
Un camión puede transportar mercancías de 5 tipos diferentes, con los datos siguientes:

Restricciones:
* Peso máximo: 112
* Volumen máximo: 109

| Tipo | Peso [t] | Volumen [m3] | Valor [EUR] |
|------|----------|--------------|-------------|
| 1    | 5        | 1            | 4           |
| 2    | 8        | 8            | 7           |
| 3    | 3        | 6            | 6           |
| 4    | 2        | 5            | 5           |
| 5    | 7        | 4            | 4           |


```python
import gurobipy as grb
```

```python
# Datos
# xj : Números de unidades de tipo j cargadas
Tipos = [1, 2, 3, 4, 5]
Peso = [5, 8, 3, 2, 7]
Volumen = [1, 8, 6, 5, 4]
Valor = [4, 7, 6, 5, 4]

Peso_Max = 112
Volumen_Max = 109

m = grb.Model("Carga de Camión")
x = m.addVars(Tipos, vtype=grb.GRB.INTEGER, name="x")

# Función objetivo
m.setObjective(grb.quicksum(
    Valor[i] * x[Tipos[i]] for i in range(len(Tipos))), grb.GRB.MAXIMIZE)
```

```python
# Restricciones

# Restricciones peso
m.addConstr(grb.quicksum(
    Peso[i] * x[Tipos[i]] for i in range(len(Tipos))
) <= Peso_Max, "Capacidad_Peso"
            )

# Restricciones Volumen
m.addConstr(grb.quicksum(
    Volumen[i] * x[Tipos[i]] for i in range(len(Tipos))
) <= Volumen_Max, "Capacidad_Volumen"
            )
print(grb.quicksum(
    Volumen[i] * x[Tipos[i]] for i in range(len(Tipos))))
print(grb.quicksum(
    Peso[i] * x[Tipos[i]] for i in range(len(Tipos))
))
```

```python
m.optimize()
if m.status == grb.GRB.OPTIMAL:
    print(f"Beneficio Máximo: {m.objVal:.2f} EUR")
```
