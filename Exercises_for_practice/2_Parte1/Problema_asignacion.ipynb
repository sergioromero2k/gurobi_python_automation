## Problema de asignaci贸n
Una empresa de alimentaci贸n dispone de cuatro ejecutivos 葛, {1,2,3,4}, y cuatro clientes 娥, {1,2,3,4}.
El coste (en millones de euros) de asignar ejecutivo  al cliente  es  , dado en la matriz de costes =().

**Tarea**: formular el programa matem谩tico que modele el problema de asignaci贸n
uno a uno que minimiza el coste total.

| Par谩metros | C1 | C2 | C3 | C4 |
| --- | --- | --- | --- | --- |
| E1 | 15 | 19 | 20 | 18 |
| E2 | 14 | 15 | 17 | 14 |
| E3 | 11 | 15 | 15 | 14 |
| E4 | 21 | 24 | 26 | 24 |

```python
# Datos
Ejecutivos = [1, 2, 3, 4]
Clientes = [1, 2, 3, 4]
Costo = {
    (1, 1): 15,
    (1, 2): 19,
    (1, 3): 20,
    (1, 4): 18,
    (2, 1): 14,
    (2, 2): 15,
    (2, 3): 17,
    (2, 4): 14,
    (3, 1): 11,
    (3, 2): 15,
    (3, 3): 15,
    (3, 4): 14,
    (4, 1): 21,
    (4, 2): 24,
    (4, 3): 26,
    (4, 4): 24,
}
```

```python
import gurobipy as grb

m = grb.Model("Problema de asignacion")
```

```python
# Variables
X = m.addVars(Ejecutivos, Clientes, vtype=grb.GRB.BINARY, name="X")

# Funci贸n Objetivo
m.setObjective(
    grb.quicksum(X[i, j] * Costo[i, j] for i in Ejecutivos for j in Clientes),
    grb.GRB.MINIMIZE,
)
```

```python
m.addConstrs(
    (grb.quicksum(X[i, j] for i in Ejecutivos) == 1 for j in Ejecutivos),
    name="ClientePorEjecutivo",
)

m.addConstrs(
    (grb.quicksum(X[i, j] for j in Clientes) == 1 for i in Ejecutivos),
    name="EjecutivoPorCliente",
)
```

```python
m.optimize()

if m.status == grb.GRB.Status.OPTIMAL:
    print("\n--- Resultados ---")
    print(f"Costo M铆nimo Total: {m.objVal:.2f}\n")
```
