## Problema de la dieta

Un veterinario quiere elaborar un compuesto que permita alimentar a los chimpanses y que cubra sus
necesidades mínimas de calcio, fósforo y magnesio al mínimo coste posible. Para ello se debe mezclar
de forma adecuada dos preparados alimenticios, A y B. Una porción de A contiene 30 mg de calcio, 10
mg de fósforo y 40 mg de magnesio, y cuesta 3 euros. Mientras que una porción de B contiene 40 mg de
calcio, 30 mg de fósforo y 20 mg de magnesio, y tiene un precio de 4 euros.
Si la dieta debe aportar, al menos 350 mg de calcio, 150 mg de fósforo y 300 mg de magnesio.

¿Cuántas porciones de cada preparado deben utilizarse para satisfacer estos requisitos con
el mínimo coste? ¿En estos casos el chimpanse recibirá más cantidad de calcio que la necesaria?

| Parámetros | Calcio | Fósforo | Magnesio | Coste (€/u) |
| --- | --- | --- | --- | --- |
| Preparado A (x1) | 30 | 10 | 40 | 3 |
| Preparado B (x2) | 40 | 30 | 20 | 4 |
| Cantidades mínimas (mg) | 450 | 150 | 300 | - |

```python
# Datos
# X1: Preparado A
# X2: Preparado B

import gurobipy as grb

model = grb.Model("Problema de la dieta")

# Variables
x1 = model.addVar(vtype=grb.GRB.CONTINUOUS, name="Prep_A")
x2 = model.addVar(vtype=grb.GRB.CONTINUOUS, name="Prep_B")

# Sujeto a: (Restricciones)
model.addConstr(30 * x1 + 40 * x2 <= 450, name="Prep_Calcio")
model.addConstr(10 * x1 + 30 * x2 >= 150, name="Prep_Fósforo")
model.addConstr(40 * x1 + 20 * x2 >= 300, name="Prep_Magnesio")

model.setObjective(3 * x1 + 4 * x2, grb.GRB.MINIMIZE)
model.optimize()
```
